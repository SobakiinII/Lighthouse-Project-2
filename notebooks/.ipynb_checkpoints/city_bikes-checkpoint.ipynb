# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import pandas as pd
import json
import requests

#Requesting the bike points for portland oregon, because it was the first city I tested that actually worked with the API
response=requests.get("https://api.citybik.es/v2/networks/biketown?fields=stations")
```

```python
print(response.json())
```

```python
#Along with the information we want, it comes with a lot of fluff, so lets trim it down to everything we need
response.json()['network']['stations']
```

```python
#with that, lets assign it a variable for comvienient parsing
working_data=response.json()['network']['stations']
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
parsed_data=[]

#using the tried and true for loop, we break this down, and append a dictionary of the information we want into a new list
for i in working_data:
    parsed_data.append({'latittude':i['latitude'],'longitude':i['longitude'], 'free_bikes':i['free_bikes']})
```

```python
print(parsed_data)
```

Put your parsed results into a DataFrame.

```python
final_dataframe=pd.DataFrame(parsed_data)
#now fully parsed, we can easily pass this into a datafrome
final_dataframe.head()
```

```python
final_dataframe.tail()
```

```python
final_dataframe.to_csv('bike_datafram.csv', index=False)
```
