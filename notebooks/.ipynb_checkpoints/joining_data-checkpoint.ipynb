Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd

bike_dataframe=pd.read_csv('bike_dataframe.csv')
yelp_dataframe=pd.read_csv('YelpDataframe.csv')
```

```python
bike_dataframe
```

```python
#To join two dataframes you need a common key between them
# I'm realizing in hindsight that the best way to acomplish this 
#would have been to attach the bike point ID to the request it recieved.

#After retroactively modifying my code, lets get started
# Join on bike id

join_df = bike_dataframe.merge(yelp_dataframe, on='id', how='inner')
```

```python
join_df.head()
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
#Number of POI to average review 

#First we make a grouped Dataframe

plot_data=join_df.groupby('id').agg({'distance':'mean', 'reviews':'mean'}).rename(columns={'distance':'avg_distance','reviews':'avg_review'}).sort_values(by = 'avg_distance')
plot_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=plot_data, x='avg_distance', y='avg_review')
plt.title('Scatter Plot of Variable X vs Variable Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3

conn = sqlite3.connect('database.db')

join_df.to_sql('joined_table', conn, if_exists='replace', index=False)
plot_data.to_sql('plot_table', conn, if_exists='replace', index=False)
conn.close()
```
